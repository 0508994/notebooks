```python
# use the whl file to install basemap 
# never use conda !!!!!
from mpl_toolkits.basemap import Basemap
import matplotlib.pyplot as plt
```

```python
plt.rcParams['figure.figsize'] = (15, 7)
```

```python
m = Basemap(projection='mill',
           llcrnrlat=-90,
           llcrnrlon=-180,
           urcrnrlat=90,
           urcrnrlon=180,
           resolution='l') # l - low, c - crude, f - full, h - high

m.drawcoastlines()
m.drawcountries(linewidth=2)
m.drawstates(color='b')
m.drawcounties(color='r')
m.etopo()
#m.fillcontinents()

plt.title('Basemap Tutorial')
plt.show
```

```python
m.drawcoastlines()
m.drawcountries(linewidth=2)
m.drawstates(color='b')
m.drawcounties(color='r')

m.bluemarble()

plt.show()
```

```python
m = Basemap(projection='mill',
           llcrnrlat=25,
           llcrnrlon=-130,
           urcrnrlat=50,
           urcrnrlon=-60,
           resolution='l') # l - low, c - crude, f - full, h - high

m.drawcoastlines()
m.drawcountries(linewidth=2)
m.drawstates(color='b')
m.drawcounties(color='r')

plt.show()
```

```python
m.drawcoastlines()
m.drawcountries(linewidth=2)
m.drawstates(color='b')
m.drawcounties(color='r')

NYClat, NYClon = 40.7127, -74.0059
xptnyc, yptnyc = m(NYClon, NYClat)
m.plot(xptnyc, yptnyc, 'c*', markersize=15)

LAlat, LAlon = 34.05, -118.25
xptla, yptla = m(LAlon, LAlat)
m.plot(xptla, yptla, 'g^', markersize=15)
m.plot([xptnyc, xptla], [yptnyc, yptla], linewidth=2, color='m', label='Flight 98') # line
m.drawgreatcircle(NYClon, NYClat, LAlon, LAlat, color='c', linewidth=2, label='Arc') # arc

plt.legend(loc=4)
plt.show()
```
